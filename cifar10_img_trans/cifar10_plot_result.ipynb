```python
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline

from cifar10_readLogFile import *
```

```python
path_fix = '/home/vicky/Documents/CaffePy/cifar10py/cifar10_caffemodels/model_modify/LOG/'
os.chdir(path_fix)
logFile_fix = ReadLogFile(path_fix)
```

## cifar-10 迭代5000次（原模型）

```python
train_iter1,train_loss1 = logFile_fix.read_trainLog('190938.log.train')
test_iter1,test_accuracy1 = logFile_fix.read_testLog('190938.log.test')
train_iter2,train_loss2 = logFile_fix.read_trainLog('191016.log.train')
test_iter2,test_accuracy2 = logFile_fix.read_testLog('191016.log.test')
train_iter = list(train_iter1[:-1])+list(train_iter2)
train_loss = list(train_loss1[:-1])+list(train_loss2)
test_iter = list(test_iter1[:-1])+list(test_iter2)
test_accuracy = list(test_accuracy1[:-1])+list(test_accuracy2)
```

```python
_, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot(train_iter,train_loss,'r-')

ax2.plot(test_iter,test_accuracy,'b-')

ax1.set_xlabel('iteration')
ax1.set_ylabel('train loss')
ax2.set_ylabel('test accuracy')
ax2.set_title('Test Accuracy: {:.4f}'.format(test_accuracy[-1]))
ax1.legend(['train loss'],loc=(0,0.9))
ax2.legend(['test accuracy'],loc=(0,0.8))
plt.show()
```
